```python
import numpy as np
from matplotlib import pyplot as pl
import os
import scipy.io.wavfile as wav

%matplotlib inline
```

```python
PATH = './vowels/'
```

```python
sample_rate_1, X_1 = wav.read(os.path.join(PATH, 'nafdlaa.wav'))
sample_rate_2, X_2 = wav.read(os.path.join(PATH, 'namshaa.wav'))
```

```python
pl.figure(figsize=(15,6))
pl.subplot(2,1,1)
pl.plot(X_1)
pl.xlim(0, len(X_1))
pl.title('X_1')
pl.grid()
pl.subplot(2,1,2)
pl.plot(X_2)
pl.xlim(0, len(X_2))
pl.title('X_2')
pl.grid()
pl.tight_layout()
```

```python
from python_speech_features import mfcc
```

```python
ceps_1 = mfcc(X_1, samplerate=sample_rate_1, nfft=1024)
ceps_2 = mfcc(X_2, samplerate=sample_rate_2, nfft=1024)
```

```python
pl.figure(figsize=(15,4))
pl.subplot(1,2,1)
pl.boxplot(ceps_1)
pl.title('X_1')
pl.grid()
pl.subplot(1,2,2)
pl.boxplot(ceps_2)
pl.title('X_2')
pl.grid()
```

# Test

```python
# src : https://stackoverflow.com/questions/3207219/how-do-i-list-all-files-of-a-directory
men_files = [file for file in os.listdir(PATH) if (os.path.isfile(os.path.join(PATH, file)) and file.startswith('nam') and file.endswith('.wav'))]
women_files = [file for file in os.listdir(PATH) if (os.path.isfile(os.path.join(PATH, file)) and file.startswith('naf') and file.endswith('.wav'))]

print("Number of natural men files =", len(men_files))
print("Number of natural women files =", len(women_files))
```

```python
men = []
for file in men_files:
    men.append(wav.read(os.path.join(PATH, file)))

women = []
for file in women_files:
    women.append(wav.read(os.path.join(PATH, file)))
    
men_mfcc = []
for (rate, X) in men:
    men_mfcc.append(mfcc(X, samplerate=rate, nfft=1024))
    
women_mfcc = []
for (rate, X) in women:
    women_mfcc.append(mfcc(X, samplerate=rate, nfft=1024))
```

```python
# test
pl.figure(figsize=(15,4))
pl.subplot(1,2,1)
pl.boxplot(men_mfcc[0])
pl.title('X_1')
pl.grid()
pl.subplot(1,2,2)
pl.boxplot(women_mfcc[0])
pl.title('X_2')
pl.grid()
```
